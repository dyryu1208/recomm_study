## Top K Recommendation

- 사용자 혹은 세션별로 Top K 추천 결과를 생성하여 추천 성능을 평가한다.
- 구해볼 지표는 다음과 같다.
    1. Precision / Recall
    2. MAP
    3. NDCG
    4. Hit Ratio
   

- 현재 주어진 무비렌즈 데이터의 경우 사용자의 영화에 대한 평점 데이터로 이루어져있다.
- 위의 평가 지표를 구하기 위해서는 사용자에게 영화가 추천되었을 때, 사용자가 이 영화에 관련이 있는지 아닌지에 대한 기준이 필요하다.
- **따라서 사용자가 영화를 선호한다는 정의는 4.0점 이상의 평가를 내린 것으로 가정한다.**
- Top K 추천 결과 생성은 앞에서 배운 Steam 로직을 사용한 평균 평점 기반 추천 로직을 사용하고, 이를 통해 지표를 직접 구해본다.

### 추천 성능 평가를 위한 test data와 predict data 생성

```python
# 먼저 학습 데이터를 train, test로 나눈다.
train, test = train_test_split(ratings_df, test_size=0.2, random_state=10)
print(train.shape)
print(test.shape)
```

```python
# 평균 평점 기반 추천을 위한 영화별 통계량 계산
movie_statistics = pd.DataFrame({
    'numUsers' : train.groupby('movie_id')['user_id'].nunique(),
    'avgRating' : train.groupby('movie_id')['rating'].mean(),
    'stdRating' : train.groupby('movie_id')['rating'].std()
    }).reset_index()

movie_statistics.head()
```

```python
# Steam Rating 구하기
movie_statistics['steamRating'] = movie_statistics['avgRating'] - (movie_statistics['avgRating'] - 3.0) \
                                  * np.power(2, -np.log10(movie_statistics['numUsers']))
topk_df = movie_statistics.sort_values(by='steamRating', ascending=False)
topk_df
```

### Precision / Recall / MAP

```python
# recommend의 길이는 @K를 의미

def get_precision(relevant, recommend):
    
    '''
    relevant : 실제로 구매한 item
    recommend : 모델이 추천한 item
    _intersection --> 추천된 것과 실제 구매한 것(relevant)의 교집합
    precision = _intersection 개수 / 모델이 추천한 item의 개수
    '''
    _intersection = set(recommend).intersection(set(relevant))
    return len(_intersection) / len(recommend)

def get_recall(relevant, recommend):
    '''
    _intersection = precision과 동일
    recall = _intersection / 실제 구매한 item의 개수(relevant)
    '''
    _intersection = set(recommend).intersection(set(relevant))
    return len(_intersection) / len(relevant)

def get_average_precision(relevant, recommend):
    '''
    MAP : item별 평균 precision
    '''
    _precisions = []
    
    for i in range(len(recommend)):
        _recommend = recommend[:i+1]
        _precisions.append(get_precision(relevant, _recommend))
    
    return np.mean(_precisions)
```

```python
# 개별 User에 대해 추천을 수행하고 각 지표를 구한 후 합친다
# @로 사용할 k = 10으로 설정

test_user_set = set(test['user_id'].unique())
k = 10
recommend_item = topk_df['movie_id'][:k].tolist()  # steamRating 기준으로 점수가 가장 높은 10개의 영화 추천

precisions = []
recalls = []
average_precisions = []

for user_id in list(test_user_set):

  test_user_rating_df = test[(test['user_id'] == user_id) & (test['rating'] >= 4.0)]  # User가 영화를 선호하는 기준 --> Rating 4.0 이상
  relevant_item = test_user_rating_df.sort_values(by='rating', ascending=False)['movie_id'].tolist()

  # test에 있는 User 중 선호 영화가 없는 케이스도 존재(4.0 이상 매긴 영화가 아예 없는 User)
  # 해당 유저는 배제
  if len(relevant_item) == 0:
    continue

  # precision@10
  precision = get_precision(relevant_item, recommend_item)
  precisions.append(precision)

  # recall@10
  recall = get_recall(relevant_item, recommend_item)
  recalls.append(recall)

  # MAP@10
  average_precision = get_average_precision(relevant_item, recommend_item)
  average_precisions.append(average_precision)

print("precision@k: ", np.mean(precisions))
print("recall@k: ", np.mean(recalls))
print("map@k: ", np.mean(average_precisions))  
```

### NDCG
* 해당 수식에서는 평점이 고려되지 않고, 단순히 샀냐(1) 안샀냐(0)으로 분류되는 듯

```python
# 특정 사용자에 대해서 5개의 아이템이 추천되었을 때,

relevant_item = ['a', 'b', 'c']
recommend_item = ['d', 'c', 'a', 'b', 'e']
k = len(recommend_item)
```

```python
# CG(cumulative gain) 구하기
cg = []
for item in recommend_item:
  if item in relevant_item:
    cg.append(1)
  
  else:
    cg.append(0)
print(cg)
```

```python
# k에 대한 discount 구하기
discount = np.log2(np.arange(k) + 2)
discount
```

```python
# DCG(Discounted CG) 구하기
dcg = np.sum(np.divide(cg, discount))
dcg
```

```python
# ICG(Ideal CG) 구하기
# User의 relevant item 개수가 k보다 적은 경우가 존재. 예를 들어, 
# 1) k = 5, 유저의 relevant item 개수가 5개 이상이라면 idcg = (1/log2) + (1/log3) + (1/log4) + (1/log5) + (1/log6)이지만,
# 2) k = 5, 유저의 relevant item 개수가 3개라면 idcg = (1/log2) + (1/log3) + (1/log4)이다.

k_for_icg = min(k, len(relevant_item))   # k와 relevant_item 중 작은 값으로 계산
icg = np.zeros(k)

for i in range(k_for_icg):
  icg[i] += 1

icg
```

```python
# ideal dcg 구하기
idcg = np.sum(np.divide(icg, discount))
idcg
```

```python
ndcg = dcg / idcg
ndcg
```

```python
# NDCG@k 함수 생성
# k개의 item이 추천되었을 때 NDCG값을 구함

def get_ndcg(relevant_item, recommend_item):

  '''
  k = 추천된 item의 수
  1) cg를 구하고, dcg 계산
  2) icg를 계산 후, 최종적으로 NDCG 계산
  '''
  k = len(recommend_item)
  discount = np.log2(np.arange(k) + 2)
  cg = []

  for item in recommend_item:
    if item in relevant_item:
      cg.append(1)
    else:
      cg.append(0)
  
  dcg = np.sum(np.divide(cg, discount))
  k_for_icg = min(k, len(relevant_item))
  icg = np.zeros(k)
  
  for i in range(k_for_icg):
    icg[i] += 1
  
  idcg = np.sum(np.divide(icg, discount))

  return dcg / idcg

relevant_item = ['a', 'b', 'd', 'f']
recommend_item = ['d', 'c', 'a', 'b', 'e']
get_ndcg(relevant_item, recommend_item)
```

```python
# 개별 사용자에 대해서 추천을 수행하고 각 지표를 구한 뒤에 이를 합친다.

k = 10

recommend_item = topk_df['movie_id'][:k].tolist()

ndcgs = []

for user_id in list(test_user_set):
    
    test_user_rating_df = test[(test['user_id'] == user_id) & (test['rating'] >= 4.0)]
    relevant_item = test_user_rating_df.sort_values(by='rating', ascending=False)['movie_id'].tolist()
    
    # 테스트 데이터에 있는 유저 가운데 선호 영화가 아예 없는 케이스도 존재함. (4.0이상 평가한 영화가 아예 없는 유저)
    if len(relevant_item) == 0:
        continue
        
    # ndcg@k
    ndcg = get_ndcg(relevant_item, recommend_item)
    ndcgs.append(ndcg)

print("ndcg@k: ", np.mean(ndcgs))
```

## Rating Prediction

- Top-K Recommendation이 아니라 사용자가 아이템에 대해 내린 선호도를 정확히 예측하는 문제로 접근해보자.
- 3가지 간단한 추천 알고리즘을 사용해 RMSE, MAE 값을 구해보자.
    1. 영화 평균 평점으로 예측
    2. 사용자 평균 평점으로 예측

```python
# train / test 데이터에 있는 사용자와 영화 리스트를 잠시 살펴보자.

# 사용자
train_user_set = set(train['user_id'].unique())
test_user_set = set(test['user_id'].unique())
print("test 데이터에 존재하지만 train 데이터에 등장하지 않는 유저: ", len(test_user_set - train_user_set))

# 영화
train_movie_set = set(train['movie_id'].unique())
test_movie_set = set(test['movie_id'].unique())
print("test 데이터에 존재하지만 train 데이터에 등장하지 않는 영화: ", len(test_movie_set - train_movie_set))
```

### 1. 영화 평균 평점 구하기

- train 데이터에 있는 평점을 사용해 영화 평균 평점을 구한다.
- test 데이터의 평점을 예측할 때 이 평균 평점을 사용한다.
- 만약 test 데이터에 등장한 영화에 대한 평균 평점이 train 데이터에 없는 경우에는 전체 평균 평점으로 예측한다.

```python
# train 데이터의 영화별 평균평점 계산
train_movie_df = train.groupby('movie_id').mean()
train_movie_df
```

```python
from collections import defaultdict

movie_rating = defaultdict(float)

for idx, row in train_movie_df.iterrows():
  movie_rating[idx] = row.rating  # movie_id는 key, 평균 평점은 value

print(movie_rating)
```

```python
total_average_rating = train['rating'].mean()
print("전체 평균: ", total_average_rating)
```

```python
# test 데이터 영화 평점 예측

test['predicted_rating_movie'] = test['movie_id'].apply(lambda x: movie_rating.get(x, total_average_rating))     # dictionary의 get 메서드 : 첫번째 인자로 넣고자 하는 값을 넣고, 없으면 2순위 값 매김
test
```

### 2. 사용자 평균 평점 구하기
- train 데이터에 있는 평점을 사용해 사용자별 평균 평점을 구한다.
- test 데이터의 평점을 예측할 때 해당 사용자의 평균 평점을 사용한다.
- 마찬가지로 test 데이터에 등장한 사용자에 대한 평균 평점이 train 데이터에 없는 경우에는 전체 평균 평점으로 예측한다.

```python
train_user_df = train.groupby('user_id').mean()
train_user_df
```

```python
from collections import defaultdict

user_rating = defaultdict(float)
for index, row in train_user_df.iterrows():
    user_rating[index] = row.rating
user_rating
```

```python
# 평점 예측하기

test['predicted_rating_user'] = test['user_id'].apply(lambda x: user_rating.get(x, total_average_rating))
test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 영화 평균 평점을 사용한 추천
movie_mae = mean_absolute_error(test['rating'].values, test['predicted_rating_movie'].values)
movie_rmse = np.sqrt(mean_squared_error(test['rating'].values, test['predicted_rating_movie'].values))

print('영화 평균 평점을 사용한 예측 오차')
print(f'MAE : {movie_mae} \t RMSE : {movie_rmse}')

# 사용자 평균 평점을 사용한 추천
user_mae = mean_absolute_error(test['rating'].values, test['predicted_rating_user'].values)
user_rmse = np.sqrt(mean_squared_error(test['rating'].values, test['predicted_rating_user'].values))

print('사용자 평균 평점을 사용한 예측 오차')
print(f'MAE : {user_mae} \t RMSE : {user_rmse}')
```
